7-Her kıta için ve her alkol cinsi için ortalama alkol tüketimini gösteriniz. 

```python
df.groupby(["continent"])[["beer_servings", "spirit_servings", "wine_servings"]].mean()
```

8-Her kıta için ve aşağıdaki alkol cinsleri için belirtilen hesaplamaları gerçekleştiriniz.

beer_servings - std
spirit_servings - median 
wine_servings - min, mean, max
total_litres_of_pure_alcohol - mean, count

```python
print(df.groupby("continent")["beer_servings"].std())
print(df.groupby("continent")["spirit_servings"].median())
print(df.groupby("continent")["wine_servings"].min())
print(df.groupby("continent")["wine_servings"].mean())
print(df.groupby("continent")["wine_servings"].max())
print(df.groupby("continent")["total_litres_of_pure_alcohol"].mean())
print(df.groupby("continent")["total_litres_of_pure_alcohol"].count())
```

9- Her alkol cinsinin en çok tüketildiği ülkeyi bulun. 

```python
print(df.sort_values(by="beer_servings", ascending=False).head(1)['country'])
print(df.sort_values(by="wine_servings", ascending=False).head(1)['country'])
print(df.sort_values(by="spirit_servings", ascending=False).head(1)['country'])
```

10- Yeni değişkenler üreterek her alkol cincisin ilgili ülkedeki toplam alkol tüketimi içerisindeki oranlarını yazdırınız.  

```python
df["beer_ratio"] = df["beer_servings"] / (df["beer_servings"] + df["wine_servings"] + df["spirit_servings"])
df["wine_ratio"] = df["wine_servings"] / (df["beer_servings"] + df["wine_servings"] + df["spirit_servings"])
df["spirit_ratio"] = df["spirit_servings"] / (df["beer_servings"] + df["wine_servings"] + df["spirit_servings"])


df
```
